# `006` Batch normalization

Requirements: 005 Multilayer perceptron.

Batch normalization (BN or BatchNorm) is a technique to make trainign faster and more stable by normalizing the input of a layer. The idea is to normalize the input of a layer by subtracting the mean and dividing by the standard deviation.

```python
from matplotlib import pyplot as plt
import torch

# let's make a simple X with a strange distribution
X = 5 * torch.rand(100)**2 - torch.cos(torch.rand(100))
plt.subplot(211)
plt.hist(X, bins=20)

# now, let's normalize it
X = (X - X.mean()) / X.std()
plt.subplot(212)
plt.hist(X, bins=20);
```
