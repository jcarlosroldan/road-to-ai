In maths, a the partial derivative of a function $f(x)$ is another function $f'(x)$ that, for the same $x$, represents the direction in which that function is growing, often called the slope or gradient. For instance, let's visualize the  gradient of the function $f(x) = -x^3 - 3x^2 + 3x$ at the point $x=0$.

```python
from matplotlib import pyplot as plt
import numpy as np

f = lambda x: -x**3 - 3*x**2 + 3*x

x = np.linspace(-2, 2, 100)
y = f(x)
y_tangent = 3*x

plt.plot(x, y, label='$f(x)$', c='r')
plt.plot(x, y_tangent, label='derivative of f at $x=0$; a.k.a $f\'(0)$', c='black')
plt.plot([0], [0], 'ko')
plt.legend();
```

How do we know that the gradient at $x=0$ is $3$? There is a intuitive way to compute it, which is the following:

$$f'(x) = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h}$$

We can approximate this using some small value for $h$ in code, let's see it:

```python
h = 0.00000001
x = 0

(f(x + h) - f(x))/h
```

The intuitive idea behind this limit is the following: if you look at a function in a given point $f(x)$, and at the same function in a point slightly greater $f(x + h)$, the difference between both is how much is $f$ changing at that point. If we divide by that value of $h$, we will get the exact gradient at that interval.

This is the basic idea that mathematicians used in the past to find all the formulas for derivatives that we know today. For instance, if you have that $f(x) = x^2$, you can replace that in the previous limit, so that:

$$f'(x) = \lim_{h \to 0} \frac{(x+h)^2-x^2}{h}$$

Now, using the binomial theorem to expand $(x+h)^2$, we get:

$$f'(x) = \lim_{h\to0} \frac{x^2+2xh+h^2-x^2}{h}$$

Notice that all the terms involving $x^2$ cancel out, and since we take the limit as $h$ approaches zero, the term $h$ disappears, and we're left with:

$$f'(x) = 2x$$

### Partial derivatives

Sometimes, we are given a function $f(x, y)$ that involves more than one parameter. In this case, we have one derivative per each parameter, since for the same $(x_0,y_0)$ increasing $x$ might make the function $f$ grow in a different direction than increasing $y$. We can still use the same trick to find the partial derivatives of the function.

$$\frac{\partial f}{\partial x} = \lim_{h\to0} \frac{f(x+h, y)-f(x, y)}{h}$$

Let's see in code the partial derivatives of the function $f(x, y) = 2x + xy - x^3 + y / 2$ at the origin:

```python
f = lambda x, y: 2*x + x*y - x**3 + y/2
x = 0
y = 0
h = 0.00000001

print(f'The derivative of f with respect to x at (0, 0) is {(f(x + h, y) - f(x, y)) / h}')
print(f'The derivative of f with respect to x at (0, 0) is {(f(x, y + h) - f(x, y)) / h}')
```
